```python
import os
import pandas as pd
```

```python
df = pd.read_excel('../datasets/original/all_job_seekers2021.xlsx')
df.head(10)
```

```python
df.drop(df.index[[0]], inplace=True)
df.columns = ['id','gender','region','city','age','edu_degree','major','gpa','gpa_from','inistitute','inserted_date','grad_date','record_from']
```

```python
df.head(10)
```

```python
df.drop(['record_from','grad_date','inserted_date','gpa_from','gpa'], axis=1, inplace=True)
df.head(10)
```

```python
df.fillna('لا يوجد', inplace=True)
df.head()
```

```python
g_male = df.loc[df['gender']=='ذكر']
g_female = df.loc[(df['gender']=='أنثى') | (df['gender']=='انثى')].replace('انثى','أنثى')
data = pd.concat([g_male,g_female])
```

```python
display(g_male.head(20))
print(f'\nNumber of male are seeking for job: {len(g_male)}k')
```

```python
display(g_female.head(20))
print(f'\nNumber of females are seeking for job: {len(g_female)}k')
```

```python
male_grad_dbmd = g_male.loc[(g_male['age'] > 22) & (g_male['age'] <=30) & (g_male['inistitute'].str.startswith("جامعة"))]
female_grad_dbmd = g_female.loc[(g_female['age'] > 22) & (g_female['age'] <=30) & (g_female['inistitute'].str.startswith("جامعة"))]
```

```python
male_grad_dbmd.head()
```

```python
female_grad_dbmd.head()
```

```python
male_grad_hs = g_male.loc[(g_male['age'] > 22) & (g_male['age'] <=30) & (g_male['edu_degree'].str.startswith("ثانوية"))]
female_grad_hs = g_female.loc[(g_female['age'] > 22) & (g_female['age'] <=30) & (g_female['edu_degree'].str.startswith("ثانوية"))]
```

```python
male_grad_hs.head()
```

```python
male_grad_hs.to_csv('../datasets/created/males_without_dbmd.csv', index=False)
female_grad_hs.to_csv('../datasets/created/females_without_dbmd.csv', index=False)
male_grad_dbmd.to_csv('../datasets/created/males_with_dbmd.csv', index=False)
female_grad_dbmd.to_csv('../datasets/created/females_with_dbmd.csv', index=False)
g_male.to_csv('../datasets/created/total_males.csv', index=False)
g_female.to_csv('../datasets/created/total_females.csv', index=False)
data.to_csv('../datasets/created/seekers.csv', index=False)
```
