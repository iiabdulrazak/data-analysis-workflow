```python
import pandas as pd
```

```python
df = pd.read_excel('../datasets/original/all_job_seekers2021.xlsx')
df.head()
```

```python
df.drop(df.index[[0]], inplace=True)
df.columns = ['id','gender','region','city','age','edu_degree','major','gpa','gpa_from','inistitute','inserted_date','grad_date','record_from']
```

```python
df.head()
```

```python
df.drop(['record_from','grad_date','inserted_date','gpa_from','gpa'], axis=1, inplace=True)
df.head(10)
```

```python
df.replace('انثى','أنثى', inplace=True)
```

```python
df.fillna('لا يوجد', inplace=True)
```

```python
print(f'all job seekers before cleaning: {len(df)}k')
```

```python
un_wanted = df[(df['gender'] == 'لا يوجد') | (df['edu_degree'] == 'لا يوجد') | (df['major'] == 'لا يوجد') | (df['inistitute'] == 'لا يوجد')].index
df.drop(un_wanted , inplace=True)
```

```python
print(f'all job seekers after cleaning: {len(df)}k')
```

```python
males = df.loc[(df['gender'] == 'ذكر')]
females = df.loc[(df['gender'] == 'أنثى')]
```

```python
print(f'males job seekers after cleaning: {len(males)}k')
print(f'females job seekers after cleaning: {len(females)}k')
```

```python
df['gender'].value_counts().plot.pie(figsize=(12,6))
```

```python
df['gender'].value_counts().plot.barh(figsize=(11,6))
```

```python
df['age'].plot.hist(figsize=(12,6))
```

```python
male_grad_dbmd = df.loc[(df['gender'] == 'ذكر') & (df['age'] > 22) & (df['age'] <=30) & (df['inistitute'].str.startswith("جامعة"))]
female_grad_dbmd = df.loc[(df['gender'] == 'أنثى') & (df['age'] > 22) & (df['age'] <=30) & (df['inistitute'].str.startswith("جامعة"))]
```

```python
males_females_with_dbmd = pd.concat([male_grad_dbmd,female_grad_dbmd])
```

```python
display(males_females_with_dbmd.head(3))
print(f'males and females between age 22 to 30 with university degree: {len(males_females_with_dbmd)}k\n')

display(male_grad_dbmd.head(3))
print(f'only males between age 22 to 30 with university degree: {len(male_grad_dbmd)}k\n')

display(female_grad_dbmd.head(3))
print(f'only fmales between age 22 to 30 with university degree: {len(female_grad_dbmd)}k')
```

```python
df.to_csv('../datasets/created/seekers.csv', index=False)
males.to_csv('../datasets/created/total_males.csv', index=False)
females.to_csv('../datasets/created/total_females.csv', index=False)
male_grad_dbmd.to_csv('../datasets/created/males_with_dbmd.csv', index=False)
female_grad_dbmd.to_csv('../datasets/created/females_with_dbmd.csv', index=False)
males_females_with_dbmd.to_csv('../datasets/created/males_and_females_with_dbmd.csv', index=False)
```
